```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stocks = pd.read_csv('techiposince2001.csv')
```

```python
stocks.dtypes
```

```python
stocks = stocks.loc[stocks['ADR TSO'] == 'n/a']
```

```python
stocks.MarketCap = stocks.MarketCap.apply(lambda x: round(x/10000))
#stocks.MarketCap = stocks.MarketCap.apply(lambda y: '${:,.0f}'.format(y))
```

```python
stocks.head()
```

```python
test =  stocks.groupby(stocks.Industry,as_index=False).count()
test2 = pd.DataFrame(test)
```

```python
test2 = test2.sort_values(by='Symbol', ascending=False)
```

```python
test2
```

```python

labels = test2['Industry']
values = test2['Symbol']
```

```python
color_dict = {'royalblue','sienna','red','gold','chartreuse','darkgreen','mediumspringgreen','darkcyan','deepskyblue','firebrick','navy','mediumpurple',
             'plum','m','silver'}
```

```python
plt.pie(values, labels=labels,radius=100.5,autopct='%.0f%%',startangle=90,colors=color_dict)
plt.axis('equal')
plt.title('Percentage of IPO\'s by Industry', loc='left', fontsize=30,va='top')
fig = plt.gcf()
fig.set_size_inches(10,10)

plt.show()
```

```python
markcap = stocks.groupby(stocks.Industry).sum()
markcap
```

```python
marketcapvalues = markcap['MarketCap']
```

```python
plt.pie(marketcapvalues, labels=labels,radius=100.5,autopct='%.0f%%',startangle=90,colors=color_dict)
plt.axis('equal')
plt.title('Market Cap of IPO\'s by Industry', loc='left', fontsize=30,va='top')
fig = plt.gcf()
fig.set_size_inches(10,10)

plt.show()
```
