```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
new_IPOdate_df = pd.read_csv('IPOdate_df_exported.csv')
new_IPOdate_df.head()
```

```python
new_IPOdate_df.rename(columns={'Unnamed: 0':'Stock #'}, inplace=True)
new_IPOdate_df
```

```python
#3 - 1, 3, 5 year 0-pivoted gain/loss for the stocks
new_IPOdate_df.plot(kind = 'bar', x='IPOyear', y='Adj Close')
plt.xlabel('IPOs')
plt.ylabel('Adjusted Close Price ($)')
plt.show()
```

```python
#plt.plot(new_IPOdate_df['Stock #'],new_IPOdate_df['Adj Close'],'ro')
#plt.show()
```

```python
industry = new_IPOdate_df.groupby('Industry')
industry_performance = industry['Adj Close'].sum()
industry_performance
industry_df = pd.DataFrame(industry_performance)
final_industry_df = industry_df.reset_index()
final_industry_df
```

```python
#4 - By subsector
final_industry_df.plot(kind='bar', x='Industry', y='Adj Close')
plt.title('IPO Adjusted Close Price by Subsector')
plt.ylabel('Adjusted Close Price ($)')
plt.xlabel('Industry')
plt.show()
```
